#  1. Mean processing time (on a weekly basis)

```python
transaction_subset['created'] = pd.to_datetime(transaction_subset['created'], errors='coerce')
transaction_subset['week']=transaction_subset['created'].dt.isocalendar().week
```

```python
transaction_subset[:5]
```

```python
transaction_subset['week'].unique()
```

```python
Mean_weekly_processing_time=transaction_subset.groupby('week').agg({'processing_time':'mean'}).reset_index().sort_values(by='processing_time', ascending=False)
Mean_weekly_processing_time.dropna(axis=0, inplace=True)
Mean_weekly_processing_time.head()
```

```python
Mean_weekly_processing_time.to_excel('Mean_weekly_processing_time.xlsx', index=None)
```

```python
Mean_weekly_processing_time.isna().mean()*100
```

```python
Week=Mean_weekly_processing_time['week']
processing_time=Mean_weekly_processing_time['processing_time']
plt.title('Mean processing time (on a weekly basis)', fontname='Candara', fontsize=24, color='orange')
plt.xlabel('Week', fontname='Candara', fontsize=15, color='green')
plt.ylabel('processing_time', fontname='Candara', fontsize=15, color='green')
plt.xticks(rotation=75)
plt.bar(Week, processing_time, color="#842535")
```

# 2. Transaction Approval Rate

```python
trans_app=transaction_subset.groupby('transaction_state').agg({'transaction_type':'size'}).reset_index().sort_values(by='transaction_type', ascending=False)[:10]
trans_app['percentage of total transactions']=trans_app['transaction_type']/15391
trans_app.rename(columns={'transaction_type':'Number of Transactions'}, inplace=True)
trans_app
```

```python
trans_app.to_excel('trans_app.xlsx')
```

```python
percentage_of_total_transactions=trans_app['percentage of total transactions']
transaction_state=trans_app['transaction_state']
plt.title('Transaction Approval Rate', fontname='Candara', fontsize=24, color='orange')
plt.xlabel('percentage of total transactions', fontname='Candara', fontsize=15, color='green')
plt.ylabel('transaction_type', fontname='Candara', fontsize=15, color='green')
plt.xticks(rotation=75)
plt.barh(transaction_state, percentage_of_total_transactions, color="#842535")
```

# 3. Credit card Transaction Rate

```python
transaction_subset['transaction_type'].unique()
```

```python
card_trans=transaction_subset.groupby('transaction_type').agg({'transaction_state':'size'}).reset_index().sort_values(by='transaction_state', ascending=False)[:10]
card_trans.rename(columns={'transaction_state':'Number of Transactions'}, inplace=True)
card_trans['Number of Transactions'].sum()
card_trans['Percentage of Transactions']=card_trans['Number of Transactions']/15390
card_trans
```

```python
card_trans.to_csv('card_trans.csv')
```

```python
transaction_type=card_trans['transaction_type']
Percentage_of_Transactions=card_trans['Percentage of Transactions']
plt.title('Credit Card Transaction Rate', fontname='Candara', fontsize=24, color='orange')
plt.xlabel('Percentage of Transactions', fontname='Candara', fontsize=15, color='green')
plt.ylabel('transaction_type', fontname='Candara', fontsize=15, color='green')
plt.xticks(rotation=75)
plt.barh(transaction_type, Percentage_of_Transactions, color="#842535")
```

# 4. Average Transactions per week

```python
Monthly_trans=transaction_subset.groupby('week').agg({'transaction_type':'count'}).reset_index().sort_values(by='transaction_type', ascending=False)
Monthly_trans.rename(columns={'transaction_type':'N0. of Transactions'}, inplace=True)
Monthly_trans['No. of Transactions']=Monthly_trans['N0. of Transactions']
Monthly_trans
```

```python
Monthly_trans.to_csv('Monthly_trans2.csv', index=None)
```

```python
Monthly_trans['No. of Transactions'].sum()/53
```

```python
Week=Monthly_trans['week']
N0_of_Transactions=Monthly_trans['N0. of Transactions']
plt.title('Monthly Transactions', fontname='Candara', fontsize=24, color='orange')
plt.xlabel('Week', fontname='Candara', fontsize=15, color='green')
plt.ylabel('N0. of Transactions', fontname='Candara', fontsize=15, color='green')
plt.xticks(rotation=75)
plt.bar(Week, N0_of_Transactions, color="#842535") 
```
